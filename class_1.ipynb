```python
def getUsernameStrength(usernames, commonWords):
    
    common = set(commonWords)
    result = []

    for name in usernames:

        if len(name) < 6:
            result.append("weak")
            continue

        if name.isdigit():
            result.append("weak")
            continue

        if name.isupper() or name.islower():
            result.append("weak")
            continue

        if name in common:
            result.append("weak")
            continue

        weak = False

        for i in range(len(name)):
            for j in range(i+1, len(name)+1):
                if name[i:j] in common:
                    weak = True
                    break
            if weak:
                break

        if weak:
            result.append("weak")
        else:
            result.append("strong")

    return result

usernames = "vansh.p"
commonwords = ['vansh','apple', 'banana']
```

```python
from collections import Counter

def mostFrequent(words):

    count = Counter(words)
    max_freq = max(count.values())

    candidates = [w for w in count if count[w] == max_freq]

    return min(candidates)
```

```python
def maxSumSubarray(arr, k):

    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        print(i)

        window_sum += arr[i]
        window_sum -= arr[i-k]

        max_sum = max(max_sum, window_sum)

    return max_sum
```

```python
def longestSubstring(s):
    seen = set()
    left = 0
    ans = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left+=1
        seen.add(s[right])
        ans = max(ans, right- left +1)

    return ans , seen

ans = "abcabcabccc"
```

```python
longestSubstring(ans)
```

```python
def faltu(arr,k):
    seen = {}
    for i,n in enumerate(arr):
        x = k-n
        if x in seen:
            return [seen[x],i]
        seen[i] = n
    return None
```
