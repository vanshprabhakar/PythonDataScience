## Join, Split, Search, Sort & Filter

```python
# joining arrays

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
combined = np.concatenate((a, b))
print(combined)
```

```python
# splitting arrays
split_data = np.array_split(combined, 3)
split_data
```

```python
split_data = np.array_split(combined, 2)
split_data
```

```python
# Searching (Boolean Condition)

sales = np.array([1200, 1500, 900, 1800])

indices = np.where(sales > 1400)
print("Indices where sales > 1400:", indices)
```

```python
# sorting
sorted_sales = np.sort(sales)
sorted_sales
```

```python
#Filtering using Boolean Masking

high_sales = sales[sales>1400]
print("High Sales:", high_sales)
```

```python
revenue = np.array([1200, 1500, 1800, 900, 2000, 2200, 1700, 1600, 2500, 3000, 2800, 2600])
revenue
```

```python
#Identify high sales: sales is greater than 2000

high_sales = revenue[revenue > 2000]
high_sales
```

```python
#Represent data in quarters

quarterly = revenue.reshape(4, 3)
quarterly
```

```python
quarterly.transpose()
```
